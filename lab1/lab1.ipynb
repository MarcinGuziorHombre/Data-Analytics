```python
import pandas as pd
import numpy as np
import seaborn as sns
import bokeh
import pandas
import numpy
import scipy
import matplotlib. pyplot as plt

%matplotlib inline
sns.set()
DATA_SRC = "Data1.csv"
```

```python
#import data to csv_file.

df = pd.read_csv(DATA_SRC, index_col=0)
```

```python
#plot all columns as timestamp

df.plot()
```

```python
#histogram as sns

df.hist()
```

```python
#create df_melt

df_melt = df.melt(value_vars=df.columns.unique().to_list())
```

```python
#create histogram for each theta

g = sns.FacetGrid(df_melt, col='variable', sharex=False)
g.map(plt.hist, 'value')
```

```python
#plot kde for each theta
df.plot.kde()
```

```python
#homework

#prepare dataset
df_h = df[df.columns.unique().to_list()[0:4]]
only_in_2018 = (df_h.index > '2018-01-01 00:00:00') & (df_h.index < '2019-01-01 00:00:00')
df_h = df_h.loc[only_in_2018]
```

```python
#plot all data

df_h.plot()
```

```python
#plot all histograms

df_h.hist()
```

```python
#plot all histograms on seaborn FaceGrid

df_h_melt = df.melt(value_vars = df_h.columns.unique().to_list())
g = sns.FacetGrid(df_h_melt, col='variable', sharex=False)
g.map(plt.hist, 'value')
```

```python
#plot all kde

df_h.plot.kde()
```
